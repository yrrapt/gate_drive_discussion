## Device Characteristics

We're using the Qorvo UF3N120140 device as an example with the following relevant parameters:

$C_{ISS} = 444 \ pF$

$C_{RSS} = 45 \ pF$

$f = 10 \ MHz$

$V_DD = 18 \ V$ (gate drive voltage used by the gate driver)

## Simulation

There are three models in the simulation.

All models can be switched between including and ignoring the Miller effect with the `.param miller=0/1` parameter.

### Voltage Drive Model Simulation



This is a simple model of the FET with $C_{GS}$, $C_{GD}$, $C_{DS}$, $R_{G}$, $R_{DS_{on}}$ modelled.

The gate drive is achieved with an ideal voltage source.

### FET Drive Model Simulation



This is a simple model of the FET with $C_{GS}$, $C_{GD}$, $C_{DS}$, $R_{G}$, $R_{DS_{on}}$ modelled.

The gate drive is created with behavourial current sources that model a FET output stage which passes from a constant current region when the FET is in saturation to a resistive region when the FET enters triode. Modelling this behaviour is important to see realistic Miller effects that match those with real gate drivers.  

### FET Drive SPICE Simulation



This uses the manufacturers supplied SPICE model with the FET drive gate drive.

The simulation had convergence issues with a voltage source drive of the SPICE model so it is not included.

## No Miller Effect

We start with the simpler case where the Miller effect is not included.

In the simulation the drain is connected to an ideal voltage source and thus $\frac{\partial V_{D}}{\partial V_{G}} = 0$.

### Proposed Simple Model

In this non-Miller case I propose the simple model of an RC network, thus:

$$ P = C V^2 f $$

$$ P = 444p \cdot 18^2 \cdot 18 = 1.44 \ W $$

### Suggested Complicated Model

This is the model that was originally suggested on Discord:

$$ ((55 \ nC  10 \ MHz  2)^2  10  2.4 \ \Omega + (15 \ A)^2  (250 \ m \Omega / 2) = 57.2 \ W $$

### Comparison With Simulation

Comparing the two models and the three simulation cases we can see below:

| Method | Power |
|:-|:-|
| Proposed Analytical            | $1.44 \ W$ |
| Suggested Analytical           | $57.2 \ W$ |
| Voltage Drive Model Simulation | $1.46 \ W$ |
| FET Drive Model Simulation     | $1.36 \ W$ |
| FET Drive SPICE Simulation     | $3.51 \ W$ |

The proposed analytical and voltage & FET driven model simulation show good agreement. The SPICE model simulation is larger than 100%, showing there are probably more parasitic capacitances in the SPICE model than are accounted for in the (analytical and simulation) model consisting of just a series resistance, $C_{GS}$ and $C_{GD}$.

The suggested analytical expression is an outlier in this case.

## With Miller Effect

### Proposed Simple Model

Now we can build a more advanced model considering the Miller effect due to a changing drain voltage which increases the apparent $C_{GD}$ capacitance.

Let's assume a constant current gate drive of $2 \ A$. There are three distinct regions:
- Region 1: charging the gate-source capacitance, $C_{GS}$ ($\approx C_{ISS}$), to the threshold voltage
- Region 2: the Miller plateu as the gate drive current charge the Miller capacitance, $C_{GD}$ ($\approx C_{RSS}$)
- Region 3: the remaining charge of $C_{GS}$ ($\approx C_{ISS}$) to the gate-drive voltage

The FET parameters are: $C_{GS} = 399 \ pF$, $C_{GD} = 45 \ pF$ and $V_{TH} = 8.7 \ V$.

#### Region 1

The charge of region 1 is voltage ramp from $0 \ V$ to $V_{TH} = 8.7 \ V$.

The time spent in this region is given by:   

$$t_1 = \frac{V_{TH} \cdot C_{ISS}}{I_{GATE DRIVE}}$$

The average power dissipated is the average voltage across the gate driver multiplied by the current and scaled by the ratio of the period:

$$\overline{P_1} = (V_{DD}-\frac{1}{2}V_{TH}) \cdot I_{GATE DRIVE} \cdot \frac{t_1}{T}$$

#### Region 2

The time spent in region 2 is defined by the voltage change at the output, the $C_{GD}$ capacitance and the charging current:

$$ t_2 = \frac{V_{DS} \cdot C_{RSS}}{I_{GATE DRIVE}} = \frac{1000 \ V \cdot 45 \ pF}{2 \ A} = 22.5 \ ns $$

The average power dissipated is the average voltage across the gate driver multiplied by the current and scaled by the ratio of the period:

$$\overline{P_2} = V_{TH} \cdot I_{GATE DRIVE} \cdot \frac{t_2}{T}$$

#### Region 3

The time spent in region 2 is defined by the voltage change at the output, the $C_{GD}$ capacitance and the charging current:

$$ t_3 = \frac{(V_{DD} - V_{TH}) \cdot C_{ISS}}{I_{GATE DRIVE}} $$

The average power dissipated is the average voltage across the gate driver multiplied by the current and scaled by the ratio of the period:

$$\overline{P_3} = \frac{1}{2} V_{TH} \cdot I_{GATE DRIVE} \cdot \frac{t_2}{T}$$

#### Cycle Power

Combining all the regions and multiplying by 2 to account for charging and discharging we obtain:

$$ \overline{P_{GATE DRIVE}} = 2 \cdot f_{SW} \cdot \left[ C_{RSS} \cdot (V_{DS} V_{TH}) + C_{ISS} \cdot (V_{DD}^2 - \frac{3}{2} V_{DD} V_{TH} +  V_{TH}^2) \right] $$

It's important to note that the series resistance and the gate drive current do not appear in this equation, as long as the gate driver is designed sufficiently to complete charging/discharging within the cycle the magnitude of the charging current and series resistance do not affect gate driver power to the first order.

Another important item is to inspect how the expression segments the key components. If calculate the power dissipation associated with the $C_{RSS}$ and $C_{ISS}$ capacitances separately we find:

$$ \overline{P_{C_{ISS}}} = 2 \cdot f_{SW} \cdot C_{RSS} \cdot (V_{DS} V_{TH}) = 7.8 \ W $$

$$ \overline{P_{C_{RSS}}} = 2 \cdot f_{SW} \cdot C_{ISS} \cdot (V_{DD}^2 - \frac{3}{2} V_{DD} V_{TH} +  V_{TH}^2) =1.46 \ W $$

Clearly the power dissipation from the reverse capacitance induced Miller plateu is domininat and we can see that in the expression. We can also clearly see that halving the drain voltage will lead to a 42% reduction in total gate driver power dissipation. That's the importance of having the dominant effect clearly and explicitly modelled. 

### Comparison With Simulation

| Method | Power |
|:-|:-|
| Proposed Analytical            | $9.26 \ W$ |
| Suggested Analytical           | $57.2 \ W$ |
| Voltage Drive Model Simulation | $9.17 \ W$ |
| FET Drive Model Simulation     | $9.08 \ W$ |
| FET Drive SPICE Simulation     | $11.8 \ W$ |

Here we see good agreement again between the proposed analytical, and both the simulation models. The SPICE simulation is also of a similiar magnitude.

Again the suggested analytical model does not seem to match the other findings.
